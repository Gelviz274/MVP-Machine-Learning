```python
import pandas as pd
import numpy as np
```

```python
# Cargar base de datos 
df = pd.read_csv("data.csv")
```

```python
# Verificar valores nulos
print(f"Valores nulos por columna:\n{df.isnull().sum()}")
```

```python
df.info()
```

```python
df
```

```python
# Verificar variables categoricas
df["Month"].value_counts()
df["VisitorType"].value_counts()
df["OperatingSystems"].value_counts()
df["Browser"].value_counts()
df["Region"].value_counts()
df["TrafficType"].value_counts()
```

```python
cols_categoricas = ['Month', 'VisitorType', 'OperatingSystems', 'Browser', 'Region', 'TrafficType']
```

```python
df = pd.get_dummies(df, columns=cols_categoricas, drop_first=True, dtype=int)
```

```python
df
```

```python
# 3. Convertimos las booleanas a 1 y 0
df['Weekend'] = df['Weekend'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)
```

```python
df.info()
```

```python
# 4. Verificamos el resultado
print(f"Columnas originales: {df.shape[1]}")
print(f"Columnas tras dummies: {df.shape[1]}")
```

```python
df.to_csv('data_clean.csv', index=False)
```
